```python
import requests
import pandas as pd
import folium
```

```python
json_url = "https://gbfs.urbansharing.com/rowermevo.pl/"
```

```python
stations_response = requests.get(json_url + "station_information.json")
stations_json = stations_response.json()['data']['stations']
stations_df = pd.DataFrame.from_dict(stations_json)
stations_df = stations_df.rename(columns={'name':'station_name'})
```

```python
mevo_map = folium.Map(prefer_canvas=True)
```

```python
def plotStations(station):
    folium.Circle(
        location=[station.lat, station.lon],
        radius=10,
        weight=1,
        popup=None).add_to(mevo_map)
```

```python
stations_df.apply(plotStations, axis=1);
```

```python
mevo_map.fit_bounds(mevo_map.get_bounds())
```
